```r
Auto = read.csv('/Users/jianjiamin/Code/R_code/Dataset/Auto.csv')
Auto$horsepower = as.numeric(as.character(Auto$horsepower))
Auto = na.omit(Auto)
dim(Auto)
summary(Auto)
```

```r
## (a)
#Auto$horsepower = as.numeric(Auto$horsepower)
fit <- lm(mpg ~ horsepower, data = Auto)
summary(fit)
```

(i) Yes, there is a relationship between horsepower and mpg as determined by testing the null hypothesis of all regression coefficients equal to zero. Since the F-statistic is far larger than 1 and the p-value of the F-statistic is close to zero we can reject the null hypothesis and state there is a statistically significant relationship between horsepower and mpg.

(ii) To calculate the residual error relative to the response we use the mean of the response and the RSE. The mean of mpg is $23.4459$. The RSE of the fit was $4.906$ which indicates a percentage error of $20.92 \%$. The $R^2$ of the fit was about $0.6059$, meaning $60.59 \% $ of the variance in mpg is explained by horsepower.

(iii) The relationship between mpg and horsepower is negative. The more horsepower an automobile has the linear regression indicates the less mpg fuel efficiency the automobile will have.

```r
## (a) (iv) 
predict(fit, data.frame(horsepower = 98), interval = "confidence")
predict(fit, data.frame(horsepower = 98), interval = "prediction")
```

```r
## (b) 
plot(Auto$horsepower, Auto$mpg, main = "Scatterplot of mpg vs. horsepower", xlab = "horsepower", ylab = "mpg", col = "blue")
abline(fit, col = "red")
```

```r
## (c)
par(mfrow = c(2, 2))
plot(fit)
```

The plot of residuals versus fitted values indicates the presence of non linearity in the data. The plot of standardized residuals versus leverage indicates the presence of a few outliers (higher than 2 or lower than -2) and a few high leverage points.

```r
## (f)
par(mfrow = c(2, 2))
plot(log(Auto$horsepower), Auto$mpg)
plot(sqrt(Auto$horsepower), Auto$mpg)
plot((Auto$horsepower)^2, Auto$mpg)
```

We limit ourselves to examining “horsepower” as sole predictor. It seems that the log transformation gives the most linear looking plot.

```r
fit4 <- lm(mpg~log(horsepower), data = Auto)
summary(fit4)
```

As we can see, from the F-statistic and p-value of the SLR model, we know that log(horsepower) appears to be statistically significant. 

Let's consider other combination. For example:

```r
fit5 <- lm(mpg~log(weight)+sqrt(horsepower)+acceleration+I(acceleration^2), data = Auto)
summary(fit5)
par(mfrow=c(2,2))
plot(fit5)
plot(predict(fit5), rstudent(fit5))
```

Apparently, log(weight), sqrt(horsepower), and acceleration^2 all have statistical significance of some sort. The residuals plot has less of a discernible pattern than the plot of all linear regression terms. The studentized residuals displays potential outliers (>3). The leverage plot indicates more than three points with high leverage.

There are two problems are observed from the above plots. The first one is that the residuals vs fitted plot indicates heteroskedasticity (unconstant variance over mean) in the model. And second one is that the Q-Q plot indicates somewhat unnormality of the residuals. So, a better transformation need to be applied to our model. 

We can also try to use log(mpg) to do the linear regression. The results is as follows:

```r
fit6<-lm(log(mpg)~cylinders+displacement+horsepower+weight+acceleration+year+origin,data=Auto)
summary(fit6)
```

```r
par(mfrow=c(2,2)) 
plot(fit6)
```

```r
plot(predict(fit6),rstudent(fit6))
```

The F -statistic is far from 1 and p-value of the model is closed to 0, which indicates that there is a relationship between the predictors and the responds. And Multiple R-squared is 0.8795, which means $87.95 \%$ the variance in log(mpg) is explained by these variables.
